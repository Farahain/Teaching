## Q3 Calculate, print and plot.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
```

### Sample 1

```python
t1 = np.linspace(1, 20, 20)

## for your reference ##
## X = P_m / (1 + exp(2 + (4*v_p/P_m)*(lambda_p -t) ))
X1 = 4.9 / (1 + np.exp(2 + (4*0.73/4.9)*(6.82 -t1) ))
G1 = 13 / (1 + np.exp(2 + (4*1.59/13)*(3.51 -t1) ))
L1 = 13.2 / (1 + np.exp(2 + (4*1.49/13.2)*(5.61 -t1) ))
A1 = 2.07 / (1 + np.exp(2 + (4*0.15/2.07)*(6.87 -t1) ))

print('t              X                G                    L                A')
print('----------------------------------------------------------------------------------')
for i in range(len(t1)):
    print(t1[i], X1[i], G1[i], L1[i], A1[i])
```

```python
plt.figure(figsize=(10,6))
plt.plot(t1,X1,t1,G1,t1,L1,t1,A1)
plt.legend(['Biomass, X', 'Cells, G', 'Lactic Acid, L', 'Acetic Acid, A'])
plt.title('Name of Sample 1')
```
